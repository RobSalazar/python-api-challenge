```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from ratelimit import limits
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).

> **HINT:** The OpenWeatherMap API only allows 60 calls per minute under their free plan. Try using `time.sleep(60)` after each set of 50 cities to avoid API errors due to large calls.

```python
#Define our URL and other variables to use below
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query = f"{url}&appid={weather_api_key}&units={units}&q="

#List containers to hold our results to be able to visualize after
date = []
clouds = []
wind = []
lat = []
lng = []
temp = []
humid = []
rcity = []
country = []
request = requests.get(query + city).json()

#This caps the speed at which api requests are made--50 calls a minute zZz
@limits(calls = 50 , period = 60)
def call_api():
    
    #Now that our sleep is active we have to itterate through api to get data to create a df below
    for city in cities:
        request = requests.get(query + city).json()
        
        try:
            date.append(request["dt"])
            clouds.append(request["clouds"]["all"])
            wind.append(request["wind"]["speed"])
            lat.append(request["coord"]["lat"])
            lng.append(request["coord"]["lon"])
            temp.append(request["main"]["temp_max"])
            humid.append(request["main"]["humidity"])
            country.append(request["sys"]["country"])
            rcity.append(city)
            print(f"We got one! {city}")
            
       #For whenever we dont find the records for a certain city     
        except:
            print(f"The one that got away...{city}")
            pass
    print("Done")
    
#call upon our decorator function above
call_api()
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Using above list containers and loop outputs to create a new df
city_df = pd.DataFrame({"City":rcity,
                   "Country":country,
                   "Latitude":lat,
                   "Longitude":lng,
                   "Date":date,
                   "Max Temperature":temp,
                   "Humidity":humid,
                   "Cloud Coverage":clouds,
                  "Wind Speed":wind})

#Placed an r in front to be able to differentiate from the original file in folder
city_df.to_csv("../output_data/cities.csv")
city_df
```

## Linear Regression

```python
#I will create new dataframes for both the northern and souther hemisphere
north = city_df.loc[city_df["Latitude"] > 0]
south = city_df.loc[city_df["Latitude"] < 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Grabbing variables for scatter plot
x = north["Latitude"]
y = north["Max Temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regres = x * slope + intercept

#Plot and format the scatter
plt.figure(figsize = (11 , 7))
plt.scatter(x , y , marker = "o" , edgecolor = "black")

#Plot the regression line and add the equation in the model
plt.plot(x , regres , "r-")
mx_b = "y = " + str(round(slope , 2)) + "x + " + str(round(intercept , 2))
plt.annotate(mx_b , (10 , 10) , fontsize = 14 , color = "red")

#Adding titles and labels added date to change dynamically whenever this is run
plt.title("Northern Hemisphere - Latitude vs Max Temperature %s" % time.strftime ("%x"), size=14)
plt.xlabel("Latitude", size = 12)
plt.ylabel("Temperature (F)" , size = 12)
plt.grid(True)
plt.show
plt.savefig("../output_data/fig5.png")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x = south["Latitude"]
y = south["Max Temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regres = x * slope + intercept

#Plot and format the scatter
plt.figure(figsize = (11 , 7))
plt.scatter(x , y , marker = "o" , edgecolor = "black")

#Plot the regression line and add the equation in the model
plt.plot(x , regres , "r-")
mx_b = "y = " + str(round(slope , 2)) + "x + " + str(round(intercept , 2))
plt.annotate(mx_b , (-50 , 90) , fontsize = 14 , color = "red")

#Adding titles and labels added date to change dynamically whenever this is run
plt.title("Southern Hemisphere - Latitude vs Max Temperature %s" % time.strftime ("%x"), size=14)
plt.xlabel("Latitude", size = 12)
plt.ylabel("Temperature (F)" , size = 12)
plt.grid(True)
plt.show
plt.savefig("../output_data/fig6.png")
```

Northern and Southern Max Temp vs. Latitude Linear Regression

From the models we can see that there is a strong negative correlation between the latitude and temperature relationship. In the north as the temperature goes down the latitude inversely goes up. A similar, albiet less significant, relationship can be viewd in the southern hemisphere. One contributing factor to this difference is that there are far less cities in the southern hemisphere to sample, nevertheless the relationship seems to follow a similar trend as the north.
